## APPLYING FUNCTIONS TO ROWS AND COLUMNS:

### TO APPLY THE FUNCTIONS TO EACH COLUMN YOU NEED TO ADD APPLY() FUNCTION: 
    FOR EG: WE NEED TO APPLY MAX FUNCTION ON EACH COLUMN()

```python
telecom_data.apply(np.max)
```

### LAMBDA FUNCTION:They can have any number of arguments but only one expression

FOR EG: IF WE NEED TO SELECT ALL STATES STARTING WITH 'T' USING LAMBDA FUNCTION.

```python
telecom_data[telecom_data['state'].apply(lambda state:state[0]=='T')]
```

IN ABOVE OUTPUT 125 ROWS STARTS WITH STATE = 'T'.

### LET'S CHECK TOTAL DAY CALLS WHICH ARE GREATER THAN 90.

```python
telecom_data[telecom_data['total day calls'] > 90]
```

### MAPPING:THIS METHOD CAN BE USE TO REPLACE A VALUE IN A COLUMN FOR EG:{OLD VALUE:NEW VALUE} AS AN ARGUMENT.

```python
F={'no':False,'yes':True}
telecom_data['international plan']=telecom_data['international plan'].map(F)
```

```python
telecom_data.head()
```

Again do these for voice mail plan:

```python
telecom_data['voice mail plan']=telecom_data['voice mail plan'].map(F)
```

```python
telecom_data.head()
```

## GROUPING METHOD:

```python
columns_to_use=['total day minutes','total eve minutes','total night minutes']
telecom_data.groupby(['churn'])[columns_to_use].describe(percentiles=[])
```

### CROSSTAB METHOD: THIS METHOD IS USE TO SEE HOW THE OBSERVATION IN OUR DATASET ARE DISTRIBUTED FOR TWO VARIABLES, FOR EG: LET'S TAKE CHURN AND INTERNATIONAL PLAN.

```python
pd.crosstab(telecom_data['churn'],telecom_data['voice mail plan'],normalize=True).round(2)
```

You can see in above output that majority of people i.e 60% people are loyal and do not use additional voice mail plan.

## PIVOT TABLE:

IN PIVOT TABLE WE USE THREE PARAMETERS i.e 'VALUES','INDEX','AGGFUNC'.

```python
telecom_data.pivot_table(['total day calls','total eve calls','total night calls'],['area code'],aggfunc='mean')
```

```python
telecom_data.pivot_table(['total day calls','total eve calls','total night calls'],['state'],aggfunc='max')
```

## ADDING COLUMNS TO A DATAFRAME:

```python
total_calls=telecom_data['total day calls']+telecom_data['total eve calls']+telecom_data['total night calls'] + telecom_data['total intl calls']
telecom_data.insert(loc=len(telecom_data.columns),column='Total calls',value=total_calls)
#we use loc function to define at which position we would like to insert our new column so here telecom_data.column means at the end.
```

```python
telecom_data.head()
```

```python
telecom_data.sort_values(by=['Total calls'],ascending=False)  ##sorting values of total calls in the descending order.
```
